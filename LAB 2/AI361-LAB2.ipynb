### Extra Challenge

```python
image2 = image.copy()
white_threshold = 220
black_threshold = 35
for i in range(image2.shape[0]):
    for j in range(image2.shape[1]):
        r,g,b = image2[i,j]
        if r>=white_threshold and g>=white_threshold and b >=white_threshold:
            image2[i,j,0] = 255
            image2[i,j,1]=0
            image2[i,j,2]=0
            
        if r<=black_threshold and g<=black_threshold and b<=black_threshold:
            image2[i,j,0]=0
            image2[i,j,1]=0
            image2[i,j,2]=255
io.imshow(image2)
io.show()
```
